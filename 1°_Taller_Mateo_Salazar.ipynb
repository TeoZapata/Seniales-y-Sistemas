#– Evaluar la expresión $\int_{-\infty}^{\infty} t^{2}e^{-\sin(t)}\cos(2t)\delta(2t-2\pi)dt$ Nota: Consultar las propiedas de selectividad y escala en el tiempo de la función impulso unitario. Comprobar el resultado en simulación con la librería SymPy.


**Evaluar la expresión**
\begin{equation}
\int_{-\infty}^{\infty} t^{2}e^{-\sin(t)}\cos(2t)\delta(2t-2\pi)dt
\end{equation}

**Propiedad de selectividad**
\begin{equation}
\int_{-\infty}^{\infty} t^{2}e^{-\sin(t)}\cos(2t)\delta(2t-2\pi)dt  =  \frac  {1}{|2|} \int_{-\infty}^{\infty} t^{2}e^{-\sin(t)}\cos(2t)\delta(t-\pi)dt
\end{equation}

\begin{equation}
\frac  {1}{|2|} \pi^{2}e^{-\sin(\pi)}\cos(2\pi) = \frac {1}{2}\pi^2 \thickapprox 4.93
\end{equation}
    

```
%matplotlib inline 
# para graficar en el entorno
import sympy as sym #importa la librería sympy
sym.init_printing() #se inicializa para mostrar en pantalla

t= sym.symbols('t', real=True)
t_1=t**2
seno = sym.sin(t)
x_1=sym.exp(-seno)
cos=sym.cos(2*t)
delta=sym.DiracDelta(2*t-2*sym.pi)
total=t_1*x_1*cos*delta
total
sym.integrate(total,(t, -sym.oo, sym.oo))
```

#– Sea $x(t)=u(t-t_0)-u(t-nt_0)-k\delta(t-mt_0)$ Determine el valor de k para el cual $\int_{-\infty}^{\infty}x(t)dt=0$: Comprobar el resultado en simulaci´on con la librer´ıa SymPy.

sea
\begin{equation}
x(t)=u(t-t_0)-u(t-nt_0)-k\delta(t-mt_0)
\end{equation}
Determine el valor de k para el cual
\begin{equation}
\int_{-\infty}^{\infty}x(t)dt=0
\end{equation}

\begin{equation}
\int_{-\infty}^{\infty}[u(t-t_0)-u(t-nt_0)-k\delta(t-mt_0)] dt
\end{equation}

\begin{equation}
\int_{-\infty}^{\infty}[u(t-t_0)-u(t-nt_0)]dt - \int_{-\infty}^{\infty}k\delta(t-mt_0)] dt
\end{equation}

\begin{equation}
\int_{t_0}^{nt_0}1- \int_{-\infty}^{\infty}k\delta(t-mt_0)] dt
\end{equation}

\begin{equation}
nt_0-t_0-k=0
\end{equation}

\begin{equation}
k=t_0[n-1]
\end{equation}


```
%matplotlib inline 
# para graficar en el entorno
import sympy as sym #importa la librería sympy
sym.init_printing() #se inicializa para mostrar en pantalla

t,a,n,k,m= sym.symbols('t a n k m', real=True)
delta_1=sym.Heaviside(t-a)
delta_2=sym.Heaviside(t-n*a)
delta_3=k*sym.DiracDelta(t-m*a)
#x_de_t=delta_1-delta_2-delta_3
delta_int=sym.integrate(delta_3,(t,-sym.oo,sym.oo))
deltas=delta_1-delta_2
unit_int=sym.integrate(1,(t,a,n*a))
total=unit_int-delta_int
k_=sym.solve(total,k)
print("K = ",k_)
```

--

para $R+jwL$  
$\vec{Z}=\sqrt{R^2 + (\omega L)^2}*\angle{\frac{wL}{R}}$

con $i(t)=I_{max}sin(wt)$ $\Rightarrow$$\vec{I}=I_{max}\angle{0^°}$  

$V_{max}\angle{\theta}$


$$
\vec{V} = \vec{Z}*\vec{I}
$$
y   $\bar{P} = \bar{V}\bar{I}^*$

$$
i(t) = 5\sin{377t} \longrightarrow \bar{I} = 5\angle{0^°}\\
\omega = 377\\
L = 0.03 H\\
R = 6 \Omega\\
Z = 6 + j(377)(0.03) = 6 + j11.31
$$

```
import numpy as np
import cmath #para trabajar con complejos y fasores

I = 5
Z = 6+11.31j
V = Z*I
N = 2

Ir = np.linspace(0, abs(I), N)
Itheta = cmath.phase(I)*np.ones(N)

Vr = np.linspace(0, abs(V), N)
Vtheta = cmath.phase(V)*np.ones(N)

ax = plt.subplot(111, projection='polar')
ax.plot(Itheta, Ir, label = 'I')
ax.plot(Vtheta, Vr, label = 'V')
ax.legend()
plt.show()
print("Vemos que para el circuito RL el voltaje esta adelantado con respecto a la corriente",(Vtheta[0])*180/3.1416,"°")
```

```
I = 5
Z = 6+11.31j
V = Z*I

P = V*np.conjugate(I)

Pr = np.linspace(0, abs(P), N)
Ptheta = cmath.phase(P)*np.ones(N)

ax = plt.subplot(111, projection='polar')
ax.plot(Ptheta, Pr, label = 'P')
ax.legend()
plt.show()
print("Se pudo comprobar que la potencia tendrá una magnitud de ",abs(P)," con un angulo en ",cmath.phase(P)*(180/cmath.pi))
```

```

%matplotlib inline
from ipywidgets import interact
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```
#crear señales
h = [1,-2,0,1,-1]
x = [-5,3,-3,0,5,6-1]
#funcion convolucion
def convgraf(h,x):
   lx = len(x)
   lh = len(h)
   M = 2*lx+lh
   xm = np.zeros((M,M))
   hm = np.r_[np.zeros((lx,1)),h,np.zeros((lx,1))]
   ym = np.zeros((M,M))
   for i in range(M-lx+1):
     #print(i)
     xm[i:i+lx,i] = np.flip(x).reshape(-1)
     ym[i,i] = xm[:,i].T.dot(hm)
   return xm, ym, hm  

def plot_conv(m,xm,ym,hm):
     plt.stem(xm[:,m],markerfmt='+',label='$x[k]$')
     plt.stem(hm,linefmt='g',markerfmt='.',label='$h[n-k]$')
     plt.stem(ym[:m,:].sum(axis=0),markerfmt='s',linefmt='r',label='$y[n]=\sum^{\infty}_{k=-\infty}x[k]h[n-k]$')
     plt.legend()
     plt.show()
     return

xm, ym, hm = convgraf(h,x)
```

```
#@interact(m=range(xm.shape[0]-len(x)+1))
@interact(m=(0, xm.shape[0]-len(x), 1))
def show_frame(m=0):
    plot_conv(m,xm,ym,hm)
plt.show()
```
